```python
import matplotlib.pyplot as plt
```

```python
# Данные веб-логов

total_records = 356981
multiselect = 183148
non_multiselect = total_records - multiselect

categories = ['Мультиселект запросы', 'Обычные запросы']
values = [multiselect, non_multiselect]

plt.figure(figsize=(10, 5))
plt.barh(categories, values, color=['skyblue', 'salmon'])
plt.xlabel('Количество записей')
plt.title('Использование мультиселект запросов')

for index, value in enumerate(values):
    plt.text(value, index, str(value))
plt.xlim(0, max(values) + 15000)

plt.savefig("multiselect_count.png", bbox_inches='tight')

plt.show()
```

```python
# Данные из логов приложения

labels = ['Мультиселект запросы', 'Обычные запросы']
sizes = [90.58, 100 - 90.58]
colors = ['salmon', 'skyblue']

plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.title('Использование мультипоиска среди пользователей')

plt.savefig("multiselect_pie_app.png", bbox_inches='tight')

plt.show()
```

```python
# Данные из логов приложения

categories = [
    'Мультипоиски (из списка)',
    'Мультипоиски (из топа)'
]
values = [2638, 5300]

plt.figure(figsize=(10, 6))
bars = plt.barh(categories, values, color=['skyblue', 'salmon'])

for bar in bars:
    plt.text(bar.get_width() + 100, bar.get_y() + bar.get_height() / 2, f'{bar.get_width()}',
             va='center', ha='left')
plt.xlim(0, max(values) + 1000)

plt.title('Количество мультипоисков (по выбору "Label")')
plt.savefig("multiselect_label_app.png", bbox_inches='tight')
plt.show()
```

```python
statuses = ['499', '301', '200']
counts = [2826, 4807, 175506]
percentages = [0.79, 1.35, 49.16]

fig, ax = plt.subplots()

bars = ax.barh(statuses, counts, color=['skyblue', 'salmon', 'red'])

# Добавление подписей
for bar, percentage in zip(bars, percentages):
    width = bar.get_width()
    label_x_pos = width - 10000 if width > 10000 else width + 500
    ax.text(width + 500, bar.get_y() + bar.get_height()/2,
            f'{width:,} ({percentage:.2f}%)',
            va='center', ha='left')

ax.set_title('Количество multiselect-запросов по статусу ответа')
plt.xlim(0, max(counts) + 60000)
plt.savefig("multiselect_status_web.png", bbox_inches='tight')

plt.show()

```
